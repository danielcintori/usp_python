###Entendendo sobre Variáveis e Atribução

### O que são Variáveis?

No mundo da programação, uma **variável** é como um recipiente ou uma caixa onde você pode guardar informações. Essas informações podem ser números, textos, ou outros tipos de dados. O legal é que o conteúdo dessa 'caixa' pode mudar ao longo do tempo!

Os nomes das variáveis NÃO podem iniciar com números

Ela pode conter caracteres, _ e números

Python é case sensitive, ou seja, NUM1 e num1 são variáveis diferentes

```python
# Lemos dessa forma:
# nome = "Rafael" (nome recebe Rafael)
# numero = 1 (numero recebe 1)
telefone = 123
num1 = 2              #INTEIRO
num2 = 5.5            #FLOAT
nome = 'Rafael'       #STRING
brasil_penta = True   #BOOLEAN
brasil_hexa = False
```

### Acessando o valor de uma variável

Podemos "abrir a caixa" e ver o que está guardado dentro de uma variável usando a função `print()`.

```python
# Vamos imprimir o valor das variáveis que declaramos:
print(num1)
print(num2)
print(nome)
print(brasil_penta)
print(brasil_hexa)
```

### Reatribuindo Valores a Variáveis



Como o nome diz, o valor de uma variável pode **variar**. Podemos mudar o que está guardado em uma variável a qualquer momento.

```python
# Antes, 'num1' tinha o valor 2. Agora, vamos mudar para 10.
print("Valor atual de num1:", num1)
num1 = 10
print('Novo valor de num1:', num1)

# Vamos mudar o nome também!
nome = 'Maria'
print('Novo nome:', nome)
```

### Operadores Matemáticos Básicos em Python

Em Python, podemos realizar todas as operações matemáticas que já conhecemos. Usamos símbolos especiais, chamados **operadores**, para fazer esses cálculos.

```python
# Vamos usar algumas variáveis para demonstrar os operadores
a = 10
b = 3

# Adição
soma = a + b
print(soma)

# Subtração
subtracao = a - b
print(subtracao)

# Multiplicação
multiplicacao = a * b
print('Multiplicação entre a e b:', multiplicacao)
# print(type(multiplicacao))

# Divisão
divisao = a / b
print('Divisão:', divisao)
# print(type(divisao))

# Divisão Inteira
divisao_inteira = a // b
print('Divisão Inteira:', divisao_inteira)

# Módulo (Resto da Divisão)
modulo = a % b
print('Resto:', modulo)

# Potenciação
potencia = a ** 2 # 10 elevado a 2
print('Potência:', potencia)
```

### Ordem das Operações (Precedência)

Assim como na matemática, Python segue uma ordem para resolver as operações. Lembre-se da regra do **PEMDAS/Parênteses, Expoentes, Multiplicação e Divisão (da esquerda para a direita), Adição e Subtração (da esquerda para a direita)**.

Podemos usar parênteses `()` para forçar uma ordem específica de operações.

```python
resultado = (5 + 10) * 2
print(resultado)
```

```python
resultado1 = 2 + 3 * 4   # Multiplicação é feita antes da adição
print(resultado1)

resultado2 = (2 + 3) * 4 # Parênteses forçam a adição primeiro
print(resultado2)

expressao = (15 - 5) / 2 + 3 ** 2
#               10   / 2 + 3 ** 2
#               10   / 2 + 9
#                  5 + 9 == 14

print(expressao) # (10) / 2 + 9 = 5 + 9 = 14
```

### Operadores Relacionais (de Comparação)

Os operadores relacionais são usados para comparar dois valores. Eles sempre retornam um valor Booleano (`True` ou `False`), indicando se a comparação é verdadeira ou falsa.

```python
a = (5 != 7)
print(a)
print(type(a))
```

```python
x = 10
y = 5
z = 10

# Operador de Igualdade (==)
print(x == y)
print(x == z)

# Operador de Diferença (!=)
print(x != y)
print(x != z)

# Operador de Maior que (>)
print (x > y)
print (x > z)

# Operador de Menor que (<)
print (x < y)

# Operador de Maior ou Igual a (>=)
print (x >= y)
print (x >= z)
# Operador de Menor ou Igual a (<=)
```

### Operadores Lógicos (`and`, `or`, `not`)

Os operadores lógicos são usados para combinar expressões Booleanas (que resultam em `True` ou `False`). Eles nos permitem criar condições mais complexas para controlar o fluxo do nosso programa. Pense neles como as conjunções na nossa língua, como "e", "ou", "não".

```python
# and = e
# or = ou
# not = não

# Vamos usar um cenário de elegibilidade para desconto como exemplo
idade = 17
aluno = True
tem_cupom = False

# Operador 'and'
# Elegível para desconto estudantil: Precisa ser aluno E ter menos de 18 anos
eligivel_estudante = aluno and (idade < 18)
print('elegivel para desconto:', eligivel_estudante)

# Desconto especial: Precisa ser aluno E ter cupom
desconto_especial = aluno and tem_cupom
print('desconto especial:', desconto_especial)

# Operador 'or'
# Desconto de boas-vindas: Precisa ser aluno OU ter cupom
desconto_boas_vindas = aluno or tem_cupom
print('desconto boas vindas:', desconto_boas_vindas)

# Acesso prioritário: É aluno OU tem mais de 60 anos
acesso_prioritario = aluno or (idade > 60)
print('acesso prioritario:', acesso_prioritario)

# Operador 'not'
# Não é aluno?
nao_aluno = not aluno
print('nao eh aluno:', nao_aluno)

# Não tem cupom?
nao_tem_cupom = not tem_cupom
print('Não tem cupom:', nao_tem_cupom)
```

### Estruturas de Repetição

Aqui nosso código vai seguir até que a nosso condição de parada seja atingida!

```python
numero1 = int(input('Digite o primeiro número: '))
numero2 = int(input('Digite o segundo número: '))

while (numero1 != numero2):
  print('Os números são diferentes!')
  numero1 = int(input('Digite o primeiro número novamente: '))
  numero2 = int(input('Digite o segundo número novamente: '))

print('Os números são iguais!')
```

```python
i = 0
while (i < 5):
  print(i)
  i = i + 1
```
