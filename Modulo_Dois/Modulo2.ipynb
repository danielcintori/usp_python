```python
# Conjuntos
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Elementos que estão em A mas não em B
print("Elementos presentes em A e ausentes em B")
print(A.difference(B))

# Remove de A1 os elementos que também estão em B
print("\nRemovendo de A1 os elementos que aparecem em B")
print(A)
print(B)
A1 = A.copy()
A1.difference_update(B)
print(A1)

# Elementos comuns aos dois conjuntos
print("\nInterseção entre A e B")
print(A.intersection(B))

# Mantém em A2 apenas os elementos que também estão em B
print("\nMantendo em A2 apenas os elementos comuns a B")
print(A)
print(B)
A2 = A.copy()
A2.intersection_update(B)
print(A2)

# True se os conjuntos não possuem elementos em comum
print("\nA e B são disjuntos?")
print(A.isdisjoint(B))

# True se todos os elementos de A estão contidos em B
print("\nA é subconjunto de B?")
print(A.issubset(B))

# True se todos os elementos de A2 estão contidos em A
print("\nA é superconjunto de A2?")
print(A)
print(A2)
print(A.issuperset(A2))

# União dos conjuntos
print("\nUnião de A e B")
print(A.union(B))
```

# Matrizes

```python
# Inicialização de matriz

matriz = [[13,1,6], [-3,-9,-8], [5, 4, 6]]
print(matriz)

matriz1 = [[0 for _ in range(3)] for _ in range(3)]
print(matriz1)
```

```python
# Acessando os elementos de uma matriz
print(matriz[0][1])
print(matriz[2][2])
print(matriz[0])
```
