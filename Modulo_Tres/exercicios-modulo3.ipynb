# Exercícios de Fixação
## Módulo 2
### Exercício 1
Crie um programa que:
1. Peça ao usuário para digitar 5 itens de uma lista de compras
2. Mostre a lista completa e numerada (1 a 5)
3. Pergunte ao usuário qual item remover (pelo número do item na lista)
4. Remova esse item e mostre a lista atualizada

### Exercício 2
Crie um programa que:
1. Peça ao usuário uma quantidade N
2. Peça para cadastrar N alunos com nome, idade e nota
3. Mostre o nome de cada aluno e sua nota
4. Calcule e mostre a média dos alunos

Dica: utilize dicionários

### Exercício 3
Você tem três receitas que precisam dos seguintes ingredientes:
- Bolo: farinha, açúcar, ovos, leite e manteiga
- Panqueca: farinha, ovos, leite e sal
- Pão: farinha, água, sal e fermento

Faça um programa que responda as seguintes perguntas:
1. Quais ingredientes aparecem em todas as receitas?
2. Quais ingredientes são únicos do bolo?
3. Liste todos os ingredientes diferentes que você precisa comprar.
4. Quais ingredientes aparecem em apenas uma das receitas?

Dica: utilize sets

## Módulo 3
### Sistema de Gerenciamento de Loja
Você foi contratado para criar um sistema completo de gerenciamento de uma pequena loja. O sistema deve ter as seguintes funcionalidades:

1. Cadastro de Produtos
   - Adicionar novos produtos com: nome, preço e quantidade em estoque
2. Gestão de Vendas
   - Registrar uma venda (produto e quantidade)
   - Atualizar o estoque automaticamente após a venda
   - Calcular o valor total da venda
3. Relatório
   - Mostrar todos os produtos cadastrados
4. Menu Principal
   - Interface para o usuário escolher qual operação fazer
   - Opções:
     - Cadastrar produto
     - Realizar venda
     - Ver relatório
     - Sair

**Lembre-se**: modularidade!
